### Cleaning Data

```python
dataframe = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
station = ["Aotizhongxin", "Changping", "Dingling", "Dongsi", "Guanyuan", "Gucheng", "Huairou", "Nongzhanguan", "Shunyi", "Tiantan", "Wanliu", "Wanshouxigong"]
type_num = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", "TEMP", "PRES", "DEWP", "RAIN", "WSPM"]
type_obj = ["wd"]
```

```python
def missing_cleaning(data):
  for i in type_num :
    data[i].interpolate(method='linear', limit_direction='forward', inplace=True)

  for j in type_obj:
    data[j].fillna(value=data[j].mode()[0], inplace=True)
```

```python
for i in dataframe:
  missing_cleaning(i)
```

```python
#Mengecek missing value
df1.isna().sum()
```

```python
def outlier_cleaning(data):
  for i in type_num :
    Q1 = (data[i]).quantile(0.25)
    Q3 = (data[i]).quantile(0.75)
    IQR = Q3 - Q1

    maximum = Q3 + (1.5*IQR)
    minimum = Q1 - (1.5*IQR)

    kondisi_lower_than = data[i] < minimum
    kondisi_more_than = data[i] > maximum

    data[i] = data[i].mask(cond=kondisi_more_than, other=maximum)
    data[i] = data[i].mask(cond=kondisi_lower_than, other=minimum)
```

```python
outlier_cleaning(df1)
```

## Exploratory Data Analysis (EDA)

```python
def group_mean (data):
  return data.groupby(by="year").agg({
      "PM2.5": ["sum","mean"],
      "PM10": ["sum","mean"],
      "SO2": ["sum","mean"],
      "NO2": ["sum","mean"],
      "CO": ["sum","mean"],
      "O3": ["sum","mean"],
  })
```

```python
def group_minmax (data):
  return data.groupby(by="year").agg({
      "PM2.5": ["min","max"],
      "PM10": ["min","max"],
      "SO2": ["min","max"],
      "NO2": ["min","max"],
      "CO": ["min","max"],
      "O3": ["min","max"],
  })
```

# Visualization & Explanatory Analysis

### Pertanyaan 1:Bagaimana perkembangan konsentrasi partikel PM2.5 dan PM10 selama satu bulan dan satu minggu terakhir di setiap stasiun pengamatan selama periode pengamatan?

```python
data_2017 = all_df[all_df["year"] == 2017]

highest_pm10_city = data_2017.groupby("station")["PM10"].mean().idxmax()
lowest_pm10_city = data_2017.groupby("station")["PM10"].mean().idxmin()

selected_cities = [highest_pm10_city, lowest_pm10_city]
selected_data_2017 = data_2017[data_2017["station"].isin(selected_cities)]

plt.figure(figsize=(12, 8))
sns.lineplot(x="datetime", y="PM10", hue="station", data=selected_data_2017, ci=None, markers=True)

plt.title("Tren Konsentrasi PM10 (2017)")
plt.xlabel("Tanggal")
plt.ylabel("Konsentrasi PM10")

table_data = selected_data_2017.groupby("station")["PM10"].mean().reset_index()
plt.table(cellText=table_data.values,
          colLabels=table_data.columns,
          cellLoc="center",
          loc="bottom",
          bbox=[0.1, -0.3, 0.8, 0.1],
          colColours=["#FFDDC1", "#FFABAB"],
          fontsize=12)

plt.show()
```

```python
data_2017 = all_df[all_df["year"] == 2017]

highest_pm25_city = data_2017.groupby("station")["PM2.5"].mean().idxmax()
lowest_pm25_city = data_2017.groupby("station")["PM2.5"].mean().idxmin()

selected_cities = [highest_pm25_city, lowest_pm25_city]
selected_data_2017 = data_2017[data_2017["station"].isin(selected_cities)]

plt.figure(figsize=(12, 8))
sns.lineplot(x="datetime", y="PM2.5", hue="station", data=selected_data_2017, ci=None, markers=True)

plt.title("Tren Konsentrasi PM2.5 (2017)")
plt.xlabel("Tanggal")
plt.ylabel("Konsentrasi PM2.5")

table_data = selected_data_2017.groupby("station")["PM2.5"].mean().reset_index()
plt.table(cellText=table_data.values,
          colLabels=table_data.columns,
          cellLoc="center",
          loc="bottom",
          bbox=[0.1, -0.3, 0.8, 0.1],
          colColours=["#FFDDC1", "#FFABAB"],
          fontsize=12)

plt.show()
```

### Pertanyaan 2: Bagaimana total konsentrasi polutan di setiap stasiun pengamatan setiap tahun selama periode pengamatan?


```python
for (data, loc) in zip(dataframe, station):
  # Group data berdasarkan tahun dan kemudian menghitung total masing-masing polutan
  grouped_data = data.groupby('year').sum()

  # Reset index untuk membuat kolom 'year' menjadi kolom biasa
  grouped_data.reset_index(inplace=True)

  # Melt dataframe untuk membuat kolom polutan menjadi baris
  melted_data = pd.melt(grouped_data, id_vars=['year'], value_vars=['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'],
                        var_name='pollutant', value_name='sum_value')

  # Plot menggunakan seaborn
  plt.figure(figsize=(10, 6))
  sns.barplot(data=melted_data, x="year", y="sum_value", hue="pollutant", errorbar=None)
  plt.xlabel('Tahun')
  plt.ylabel('Total konsentrasi (ug/m^3)')
  plt.title('Total Polutan per tahun di ' + loc)
  plt.legend(loc='upper right')
  plt.show()
```

Untuk PM2.5, stasiun udara Dingling juga mencatatkan tingkat tertinggi, sementara Wanshouxigong memiliki tingkat terendah pada tahun 2017. Grafik menunjukkan bahwa Dingling memiliki konsentrasi PM2.5 yang lebih tinggi secara umum, dengan fluktuasi yang mungkin disebabkan oleh faktor lingkungan dan aktivitas manusia. Wanshouxigong, di sisi lain, menunjukkan tren yang lebih stabil dan konsentrasi PM2.5 yang relatif lebih rendah. Perbedaan ini mencerminkan variasi dalam kualitas udara antara dua stasiun tersebut, dengan konsekuensi potensial terkait kesehatan masyarakat.

Kadar polutan tertinggi di setiap stasiun selama setiap tahun umumnya terkait dengan karbon monoksida (CO), sementara sulfur dioksida (SO2) memiliki konsentrasi terendah. Analisis total polutan per tahun dapat menjadi dasar evaluasi kebijakan yang ada atau pengembangan kebijakan baru terkait polusi udara untuk meningkatkan kualitas hidup masyarakat.
